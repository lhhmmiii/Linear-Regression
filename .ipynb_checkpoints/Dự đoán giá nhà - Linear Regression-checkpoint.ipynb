# Dự đoán giá nhà bằng Linear Regerssion

**Bài toán:** Dự đoán giá nhà qua các dữ liệu thu thập là một bảng dữ liệu csv lưu các thông tin cần thiết như: diện tích nhà, diện tích phòng, thu nhập khu vực, tuổi ngôi nhà.

**Mô hình máy học:** Linear Regression.

## Import thư viện

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("USA_Housing.csv")
df=df.drop('Address',axis=1)
df.head()
```

* Avg. Area House Age: Trung bình tuổi của một ngôi nhà đã bán
* Avg. Area Number of Rooms: Trung bình diện tích các phòng
* Avg. Area Number of Bedrooms: Trung bình diện tích phòng ngủ
* Area Population: Dân số tại khu vực bán nhà
* Price: Giá ngôi nhà đã bán

## Training and Testing Data

**Lấy dữ liệu từ DataFrame**

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

**Sử dụng model_selection.train_test_split từ sklearn để chia dữ liệu thành các tập huấn luyện và thử nghiệm**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## Training the Model

**Import LinearRegression from sklearn.linear_model**

```python
from sklearn.linear_model import LinearRegression
```

**Tạo 1 instance của LinearRegression có tên lm**

```python
lm=LinearRegression()
```

**Tiến hành train dữ liệu bằng phương thức fit()**

```python
lm.fit(X_train, y_train)
```

## Predicting Test Data

**Sử dụng lm.predict() để dự đoán tập X_test.**

```python
pred=lm.predict(X_test)
```

**Sử dụng biểu đồ phân tán (scatter) truyền vào 2 giá trị trên vào để quan sát**

```python
plt.scatter(y_test,pred);
plt.xlabel('Y Test');
plt.ylabel('Predicted Y');
```

## Evaluating the Model

**Sử dụng sklearn.metrics để đánh giá mô hình**

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

* Theo chỉ số RMSE thì độ lệch giữa dự đoán và thực tế là 100,5 USD.

## Conclusion

**Độ chính xác của mô hình**

```python
metrics.explained_variance_score(y_test, pred)
```

**Hệ số coff**

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
